```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('movie-recommendation/Scripts/datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('movie-recommendation/Scripts/datasets/tmdb_5000_credits.csv')
```

```python
movies.head(3)
```

```python
credits.head(3)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(3)
```

```python
movies.info()
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(objdata):
    L = []
    for i in ast.literal_eval(objdata):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)	
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert_dict(objdata):
    L = []
    counter = 0
    for i in ast.literal_eval(objdata):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert_dict)
```

```python
movies.head()
```

```python
def fetch_director(objdata):
    L = []
    for i in ast.literal_eval(objdata):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'].apply(lambda x: ''.join(x)) +' '+ movies['genres'].apply(lambda x: ' '.join(x))+' '+movies['keywords'].apply(lambda x: ' '.join(x)) +' '+ movies['cast'].apply(lambda x: ' '.join(x))+' '+ movies['crew'].apply(lambda x: ' '.join(x))
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
new_df.loc[:, 'tags'] = new_df['tags'].apply(lambda x: x.lower())
```
